```python
import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2,3,6,4]

df = pd.DataFrame(data = {'course_name' : course_name, 'duration' : duration})
```

```python
print(df)
print(df.loc[1])
```

`iloc`, on the other hand, is used to access data by integer-based position. It refers to the location based on the numerical positions of rows and columns. Think of it like using row and column numbers (starting from 0) to find and access specific data in a table.

```python
print(df.iloc[1,1])
```

`loc` is used to access data by label or index. It refers to the location based on the row and column labels. Think of it like using the names of rows and columns to find and access specific data in a table.

```python
print(df.loc[1,'course_name'])
```

```python
index = df.index
print(index)
new_index = [3,0,1,2]
new_df = df
new_df.index = new_index
```

```python
print(new_df)
print('\n')
print(new_df.loc[2])
print('\n')
print(new_df.loc[3])
```
