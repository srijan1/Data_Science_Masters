# BOKEH 

A1. To create a Bokeh plot using Python code:
- Import the necessary modules: `from bokeh.plotting import figure, show`
- Create a figure object: `p = figure()`
- Add data and customize the plot using various glyph methods: `p.circle(x, y, size=10, color='red')`
- Show the plot: `show(p)`


A2. Glyphs in Bokeh:
- Glyphs are visual markers used to represent data points on a Bokeh plot.
- Examples of glyphs include circles, squares, lines, bars, and more.
- Glyphs can be added to a Bokeh plot using glyph methods like `circle()`, `line()`, `rect()`, etc.
- For example, to add circles to a plot: `p.circle(x, y, size=10, color='red')`


A3. Customizing the appearance of a Bokeh plot:
- You can customize various aspects of a Bokeh plot using properties and methods of the figure object.
- To customize the axes, you can set properties like `x_axis_label`, `y_axis_label`, `x_axis_type`, `y_axis_type`, etc.
- To customize the title, you can set the `title` property of the figure object.
- To add a legend, you can use the `legend` parameter in glyph methods or customize the legend using `p.legend`.


A4. Bokeh server and creating interactive plots:
- Bokeh server allows creating interactive plots that can be updated in real time.
- It provides a server-side framework for deploying Bokeh plots and applications.
- You can define callbacks and update the plot data or properties based on user interactions or streaming data.
- To use the Bokeh server, import the necessary modules, define the plot and callbacks, and run the server using `bokeh serve app.py`.


A5. Embedding Bokeh plots into web pages or dashboards:
- To embed a Bokeh plot in a web page or dashboard, you can use frameworks like Flask or Django.
- In Flask, you can define a route and render a template that includes the Bokeh plot component.
- In Django, you can create a view that renders an HTML template with the Bokeh plot.
- Both frameworks allow passing data and customizing the plot within the web application structure.
