```python
import seaborn as sns
```

Seaborn is a popular data visualization library in Python that is built on top of Matplotlib. It provides a high-level interface for creating informative and visually appealing statistical graphics. Here are five common plots in Seaborn along with their uses:

1. Scatter Plot:
   - Use: Scatter plots show the relationship between two variables as individual data points.
   - Syntax: `seaborn.scatterplot(x, y, data)`

2. Line Plot:
   - Use: Line plots display the relationship between two variables as a line connecting the data points.
   - Syntax: `seaborn.lineplot(x, y, data)`

3. Bar Plot:
   - Use: Bar plots represent categorical data with rectangular bars, where the length of each bar corresponds to the value of the category.
   - Syntax: `seaborn.barplot(x, y, data)`

4. Histogram:
   - Use: Histograms display the distribution of a single numerical variable by dividing it into bins and showing the frequency of observations in each bin.
   - Syntax: `seaborn.histplot(data, x)`

5. Box Plot:
   - Use: Box plots provide a visual summary of the distribution of a numerical variable, showing the median, quartiles, and potential outliers.
   - Syntax: `seaborn.boxplot(x, y, data)`

These are just a few examples of the many plot types available in Seaborn. Each plot serves a specific purpose and can be customized further to enhance the visual representation of the data.

```python
# Loading the dataframe
fmri = sns.load_dataset('fmri')

fmri.head()
```

```python
sns.relplot(x="timepoint", y="signal",kind="line", data=fmri)
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.boxplot(x="pclass", y="age", data=titanic,palette="ch:rot=-.25,hue=1,light=.75")
```

```python
sns.boxplot(x="pclass", y="fare", data=titanic,palette="ch:rot=-.25,hue=1,light=.75")
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
sns.histplot(data = diamonds,x = 'price')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris,hue = 'species')
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

sns.heatmap(flights_pivot)
```
