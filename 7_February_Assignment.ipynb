```python
import re
```

```python
s = 'PAssword@123'
r = 'password@123'
```

```python
def upper_lower(passwd):
    cond1 = 0
    l = '[a-z]'
    u = '[A-Z]'

    uppercase = re.findall(u,passwd)
    lowercase = re.findall(l,passwd)
    if len(uppercase)>=2 and len(lowercase)>=2:
        cond1 = 1
    return cond1
        
```

```python
upper_lower(s)
```

```python
def num_special_char_check(passwd):
    cond2 = 0
    sp = re.compile('\W')
    n = re.compile('\d')
    if len(n.findall(passwd))>=1 and len(sp.findall(passwd))==3:
        cond2 = 1
    return cond2
```

```python
num_special_char_check(s)
```

```python
def check_passwd_len(passwd):
#     cond3 = 0
    if len(passwd)>=10:
        cond3 =1
    return cond3
        
```

```python
check_passwd_len(s)
```

```python
def passwd_checker(passwd):
    cond1 = upper_lower(passwd)
    cond2 = num_special_char_check(passwd)
    cond3 = check_passwd_len(passwd)
    print(cond1,cond2,cond3)
    if cond1 == 1 and cond2 == 1 and cond3 == 1:
        return 'Valid password'
    else:
        return 'Invalid password'
```

```python
print(passwd_checker(s))
print(passwd_checker(r))
```
