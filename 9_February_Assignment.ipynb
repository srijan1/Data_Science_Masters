```python
'''Multiple inheritance is when a derived class inherits from more than one parent class. In this type of inheritance, 
the derived class has access to all the properties and methods of all its parent classes.
'''
class Parent1:
    def func1(self):
        print("This function is in parent1 class.")
        
class Parent2:
    def func2(self):
        print("This function is in parent2 class.")
        
class Child(Parent1, Parent2):
    def func3(self):
        print("This function is in child class.")
        
object = Child()
object.func1()
object.func2()
object.func3()
```

```python
'''
In Python, getter and setter methods are used to control access to class attributes or properties. 
They allow you to define custom behavior when getting (reading) or setting (assigning) the value of an attribute. 
This is useful when you want to add validation, perform calculations, or enforce certain rules before allowing access 
to the attribute.

Here's an example of a class with a getter and a setter method in Python:
'''

class Rectangle:
    def __init__(self, width, height):
        self._width = width
        self._height = height

    def get_width(self):
        return self._width

    def set_width(self, value):
        if value > 0:
            self._width = value
        else:
            print("Width must be greater than 0.")

    def get_height(self):
        return self._height

    def set_height(self, value):
        if value > 0:
            self._height = value
        else:
            print("Height must be greater than 0.")

    def get_area(self):
        return self._width * self._height

```

```python
'''
In the example above, the `Rectangle` class has two attributes: width and height. 

The getter methods (`get_width` and `get_height`) are used to retrieve the values of these attributes, 
while the setter methods (`set_width` and `set_height`) are used to assign new values to them.

In the setter methods, we can add validation logic to ensure that the new values are valid. 
In this case, we're checking that the values are greater than 0. If the validation fails, an error message is printed.
'''

rectangle = Rectangle(5, 10)
print(rectangle.get_width())  
print(rectangle.get_height()) 

rectangle.set_width(8)
rectangle.set_height(15)

print(rectangle.get_width())  
print(rectangle.get_height()) 
print(rectangle.get_area())
```

```python
'''
Method overriding is a feature in object-oriented programming that allows a subclass to provide its own implementation 
of a method that is already defined in its superclass. It is a way to modify or extend the behavior of an inherited method.

When a subclass overrides a method, it provides a new implementation of that method, which will be used instead of 
the implementation in the superclass when the method is called on an instance of the subclass.

Here's an example of method overriding in Python:
'''
class Animal:
    def sound(self):
        print("Animal makes a sound.")


class Dog(Animal):
    def sound(self):
        print("Dog barks")


class Cat(Animal):
    def sound(self):
        print("Cat meows")


# Create instances of the classes
animal = Animal()
dog = Dog()
cat = Cat()

# sound method on each instance
animal.sound()  
dog.sound()  
cat.sound() 
```
