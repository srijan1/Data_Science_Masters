```python
import numpy as np
list_ = ['1','2','3','4','5']
array_list = np.array(object = list_)
```

'list_' is of list type data str , array_list is of numpy array 
Lists and NumPy arrays are both data structures used in Python, but they have some key differences in terms of functionality and performance. Here are the main differences between lists and NumPy arrays:

1. Functionality:
   - Lists: Lists in Python are versatile and can store elements of different data types. They support dynamic resizing, meaning you can add, remove, or modify elements freely. Lists also offer a wide range of built-in methods for manipulation, such as append(), pop(), and sort().
   - NumPy Arrays: NumPy arrays are homogeneous, meaning they can only store elements of the same data type. They offer more advanced numerical computing functionalities, including element-wise operations, vectorized computations, and mathematical functions. NumPy arrays have a fixed size upon creation, and their dimensions are specified by a shape attribute.

2. Performance:
   - Lists: Lists are implemented as dynamic arrays, which means their elements are scattered throughout memory. This can lead to slower performance for operations that involve large datasets or require numerical computations.
   - NumPy Arrays: NumPy arrays are implemented as contiguous blocks of memory, allowing for efficient storage and retrieval of elements. They are optimized for numerical operations and can perform computations on entire arrays rather than individual elements, resulting in faster execution times.

3. Memory Efficiency:
   - Lists: Lists in Python can have variable lengths and may require additional memory due to their dynamic resizing nature. Each element in a list also carries additional overhead for storing type information.
   - NumPy Arrays: NumPy arrays have a fixed size and require less memory compared to lists. They store elements in a more compact and efficient manner, resulting in reduced memory usage.


In summary, lists offer more flexibility and support for diverse data types and dynamic operations, while NumPy arrays provide enhanced numerical computing capabilities, improved performance, and efficient memory usage. The choice between lists and NumPy arrays depends on the specific requirements of the task, the need for numerical computations, and the size and nature of the data being manipulated.
