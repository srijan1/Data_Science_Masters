Sure, here are 5 functions of Pandas library with execution:

1. **`read_csv()`**

This function is used to read a CSV file into a Pandas DataFrame. For example, the following code will read the `data.csv` file into a DataFrame called `df`:

```python
import pandas as pd

df = pd.read_csv('data.csv')
```

2. **`info()`**

This function provides a concise summary of a Pandas DataFrame, including the number of rows, columns, data types, and memory usage. For example, the following code will print the information about the `df` DataFrame:

```python
df.info()
```

3. **`describe()`**

This function provides a statistical summary of a Pandas DataFrame, including the mean, standard deviation, minimum, maximum, and quantiles. For example, the following code will print the statistical summary of the `df` DataFrame:

```python
df.describe()
```

4. **`value_counts()`**

This function counts the number of occurrences of each unique value in a Pandas DataFrame column. For example, the following code will count the number of times each name appears in the `name` column of the `df` DataFrame:

```python
df['name'].value_counts()
```

5. **`sort_values()`**

This function sorts a Pandas DataFrame by the values in a particular column. For example, the following code will sort the `df` DataFrame by the values in the `age` column:

```python
df.sort_values('age')
```


````python
def reindex_df(df):
    new_index = list(range(1, len(df) * 2, 2))
    df.index = new_index
    return df
````


This function takes a Pandas DataFrame `df` as input and returns a new DataFrame with the index re-indexed. The new index starts from 1 and increments by 2 for each row.

The function first creates a new list of index values, `new_index`, starting from 1 and incrementing by 2 for each row. Then, it sets the index of the DataFrame `df` to the new list of index values. Finally, it returns the re-indexed DataFrame.


````python

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})

print(df)

reindexed_df = reindex_df(df)

print(reindexed_df)
````
